```python
def copy(original):
    res=PIL.Image.new(original.mode,original.size)
    width,height=original.size
    for i in range(0,width):
        for j in range(0,height):
            pix=original.getpixel((i,j))
            res.putpixel((i,j),pix)
    return res
copied=copy(im)
copied.show()
```

```python
def rotateCC(original):
    n,m=original.size
    res=PIL.Image.new(original.mode,(m,n))
    for i in range(0,n):
        for j in range(0,m):
            pixel=original.getpixel((i,j))
            res.putpixel((j,n-1-1),pixel)
    return res
rotatedCC=rotateCC(im)
rotatedCC.show()
```

```python
#cs26
from PIL.Image import *
import PIL.Image
def crop(original,box):
    x1,y1,x2,y2=box
    width,height=x2-x1,y2-y1
    res=PIL.Image.new(original.mode,(width,height))
    for i in range(width):
        for j in range(height):
            pix=original.getpixel((x1+i,y1+j))
            res.putpixel((i,j),pix)
    return pix
```

```python
#cs28
from PIL import Image
from PIL import ImageFilter
img = Image.open("C:/Users/ANC/Desktop/download.jpg");
img.show()
def negative():
    for i in range(0, img.size[0]-1):
        for j in range(0, img.size[1]-1):
            pixelColorVals = img.getpixel((i,j));        
            redPixel    = 255 - pixelColorVals[0]
            greenPixel  = 255 - pixelColorVals[1]
            bluePixel   = 255 - pixelColorVals[2]
            img.putpixel((i,j),(redPixel, greenPixel, bluePixel))
img.show()
img.save("neg.png")
```

```python
#cs30
from PIL import Image
def rotate180(image_path, degrees_to_rotate, saved_location):
    image_obj = Image.open(image_path)
    rotated_image = image_obj.rotate(degrees_to_rotate)
    rotated_image.save(saved_location)
    rotated_image.show()
if __name__ == '__main__':
    image = "C:/Users/ANC/Desktop/download.jpg"
    rotate180(image, 180, "C:/Users/ANC/Desktop/download3.jpg")
```

```python
#cs31
from PIL import Image
def black_and_white(input_image_path,output_image_path):
    color_image = Image.open(input_image_path)
    bw = color_image.convert('L')
    bw.save(output_image_path)
if __name__ == '__main__':  
    black_and_white("C:/Users/ANC/Desktop/download.jpg",
        'bw_montana.jpg')
```
